```python
# Create map of Los Angeles
# map3 = folium.Map(location=[34.0522, -118.2437], zoom_start=10)

# Census Tracts
geo_path = '/Users/isaacmoore/Desktop/lahsatracts.geojson'

#Let Folium determine the scale
map = folium.Map(location=[34.0522, -118.2437], zoom_start=10)#, tiles='Stamen Toner')
map.choropleth(geo_path=geo_path,
               data=data_16,
               columns=['TRACTCE', 'totPeople'],
               key_on='feature.properties.TRACTCE',
               fill_color='BuGn', fill_opacity=0.7, line_opacity=0.2, line_color='black',threshold_scale=[0.1,0.3,0.5,0.7,0.9],
               legend_name='Homelessness')

###################################################
# Merge on 'censusTract' from data_16 and 'TRACTCE'
# from the geo_json file using Map.choropleth
# might have to rename 'censusTract' in data_16
###################################################

def color(beds):
    zero_25 = range(0,int(shelters.Beds.quantile(.25)))
    twentyfive_50 = range(int(shelters.Beds.quantile(.25))+1,int(shelters.Beds.quantile(.5)))
    fifty_75 = range(int(shelters.Beds.quantile(.5))+1,int(shelters.Beds.quantile(.75)))
    seventyfive_max = range(int(shelters.Beds.quantile(.75))+1, int(shelters.Beds.max()))
    col = 'white'
    if beds in zero_25:
        col='red'
    elif beds in twentyfive_50:
        col='orange'
    elif beds in fifty_75:
        col='green'
    elif beds in seventyfive_max:
        col='blue'
    return col
# Shelter Points
shelter_pt = folium.FeatureGroup(name='Shelters')
for lat, lon, name, beds in zip(shelters['X'], shelters['Y'], shelters['ProgName'], shelters['Beds']):
    shelter_pt.add_child(folium.Marker(location=[lon, lat], 
                                       popup=(folium.Popup(name)), 
                                       icon=folium.Icon(color=color(beds), icon_color='white')))
map.add_child(shelter_pt)

# Control Layer
map.add_child(folium.LayerControl())

# Save the map to a html file
map.save('shelters.html')

# Display the map
map
```
